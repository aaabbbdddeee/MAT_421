*Interpolation Problem Statement*

```python
import matplotlib.pyplot as plt
plt.style.use('seaborn-poster')

x = [0.5,2,3.5]
y = [4,2.5,1]
plt.figure(figsize = (10,8))
plt.plot(x,y,'bo', 3,4.25,'ro') 
plt.title('Data Points for Interpolation')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

*Linear Interpolation*

```python
from scipy.interpolate import interp1d 
import matplotlib.pyplot as plt
plt.style.use('seaborn-poster')

x = [0.5,1,2.4,3.7]
y = [4,2.5,2.5,6.1]
f = interp1d(x,y)
y_inter = f(1.5)
print(y_inter)    # Prints 2.75

plt.figure(figsize = (10,8))
plt.plot(x,y,'-ob',2,y_inter,'ro') 
plt.title('Linear Interpolation at x = 2')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

*Cubic Spline Interpolation*

```python
from scipy.interpolate import CubicSpline
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('seaborn-poster')

x = [1,2,3,4]
y = [6.5,3,2.5,2]
f = CubicSpline(x,y, bc_type = 'natural')
x_new = np.linspace(0,5,100)
y_new = f(x_new)

plt.figure(figsize = (10,8))
plt.plot(x_new,y_new,'b',x,y,'ro') 
plt.title('Cubic Spline Interpolation')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
     
```
